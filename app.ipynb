# Expanção de colunas

```python
import json
```

```python
columns = [] # Lista para armazenar colunas expandidas
columns_to_expand = ['trafficSource', 'device', 'totals', 'geoNetwork'] # Seleção de colunas à serem expandidas

#=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=
# Loop:
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# Para cada linha presente na coluna dentro da lista de colunas à serem expandidas, converta a coluna em um 
# dicionário, transforme-o em um DataFrame Pandas e coloque dentro de uma lista já criada chamada de coluna
#--------------------------------------------------------------------------------------------------------------
for i in columns_to_expand: 
    columns.append(pd.DataFrame([json.loads(linha) for linha in df[i]]))
#=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=/=\=

# Cocatenação de todas as colunas na lista que foi criada para junção em um único DataFrame
df_expanded = pd.concat(columns, axis = 'columns')

df[df_expanded.columns] = df_expanded # Inclusão de colunas expandidas para o DataFrame
df # Checando o DataFrame resultante
```

```python
df.drop(columns=columns_to_expand, inplace=True, errors = 'ignore') # Drop de colunas que já foram expandidas
df.head(5)
```

# Limpeza de dados

## Seleção de atributos relevantes

```python
for i in df.columns:
    tamanho_i = len(df[i].unique())
    print(f'{i}: {tamanho_i}')
```

```python
unique_values = []
columns_name = []
for i in df.columns:
    unique_values.append(len(df[i].unique()))
    columns_name.append(i)
df_n_unique_values = pd.DataFrame(dict({'column_name':columns_name, 'n_unique_values':unique_values}))
df_n_unique_values.sort_values('n_unique_values', ascending=False)
```

```python
only_1_unique_value = df_n_unique_values.loc[df_n_unique_values['n_unique_values'] == 1]['column_name'].to_numpy()
df.drop(columns=only_1_unique_value, inplace=True, errors='ignore')
df
```

## Ajuste de tipos de dados

```python
df.dtypes
```

```python
var_quant = ['bounces', 'hits', 'newVisits', 'transactionRevenue', 'pageviews']
for i in var_quant:
    df[i] = pd.to_numeric(df[i])

df.dtypes
```

# Ajuste de escala de _transactionRevenue_

```python
1 - (df.transactionRevenue.isna().sum() / df.shape[0])
```

```python
df.transactionRevenue.fillna(0, inplace=True)
```

```python
df.transactionRevenue = df.transactionRevenue / 10e5
```

# Feature Engineering

```python
df_sum = df.groupby('fullVisitorId', as_index=False)[var_quant].sum()
df_sum.head()
```

# Treinamento de baseline

## Split de dados

```python
y = df_sum.transactionRevenue.copy()
X = df_sum.drop(columns=['transactionRevenue'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
SEED = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=SEED)
```

## Treinamento

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X=X_test)
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = y_pred
resultados['erro'] = y_pred - y_test
```

```python
resultados[resultados.revenue > 0]
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
np.sqrt(mean_squared_error(y_pred=y_pred, y_true=y_test))
```
